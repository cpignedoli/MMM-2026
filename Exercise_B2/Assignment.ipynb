## Energy extraction from QE App calculations

A convenient way to organize the work is:
1. run the calculations in the **QE App**,
2. note down the **PK** of each calculation,
3. use the few cells below to extract the final energies.

Here, **PK** means the **PK of the QE App calculation**.


```python
%load_ext aiida
%aiida
```

```python
# Insert here the PK of each QE App calculation.
calculations={'4HSiC':2722,'2HSiC':2621,'3CSiC':2495,'silicon':2208,'graphite':2306,'diamond':1827,'Siatom':2149,'Catom':2120}
```

```python
# For each system:
# - load the QE App calculation from its PK
# - read the input structure
# - find the corresponding PwRelaxWorkChain
# - print final energy and chemical formula
for system in calculations:
    qe=load_node(calculations[system])
    structure=qe.inputs.structure
    relax = next(
        (wc for wc in qe.called_descendants if wc.process_label == 'PwRelaxWorkChain'),None,
    )
    print(qe.label)
    energy = relax.outputs.output_parameters.get_dict()['energy']
    print(f"Energy in eV: {energy}")
    print(f"Chemical formula: {structure.get_formula()}")
```
