## Big Drill MMM 2026 Installations
### Installing lammps with reaxff support. 

```python
import subprocess

def run(cmd):
    process = subprocess.Popen(
        cmd, shell=True, executable="/bin/bash",
        stdout=subprocess.PIPE, stderr=subprocess.STDOUT
    )
    for line in process.stdout:
        print(line.decode(), end="")
    process.wait()
    return process.returncode
```

```python
run("""
conda create -n lmp-reaxff python=3.10 -y &&
source /opt/conda/etc/profile.d/conda.sh && conda activate lmp-reaxff &&
mamba install -y -c conda-forge \
    cmake \
    make \
    gcc \
    gxx \
    openmpi \
    plumed
""")
```

```python
run("""
rm -Rf /home/jovyan/lammps-reaxff/ &&
cd $HOME &&
git clone https://github.com/lammps/lammps.git lammps-reaxff &&
source /opt/conda/etc/profile.d/conda.sh && conda activate lmp-reaxff &&
cd ~/lammps-reaxff && mkdir -p build && cd build &&
cmake ../cmake \
  -D CMAKE_BUILD_TYPE=Release \
  -D BUILD_MPI=ON \
  -D PKG_REAXFF=ON \
  -D PKG_KSPACE=ON \
  -D PKG_EXTRA-DUMP=ON \
  -D PKG_PLUMED=ON \
  -D PLUMED_MODE=static \
  -D CMAKE_INSTALL_PREFIX=$CONDA_PREFIX
""")
```

```python
run("""
source /opt/conda/etc/profile.d/conda.sh && conda activate lmp-reaxff &&
cd ~/lammps-reaxff/build &&
make -j$(nproc) &&
make install 
""")
```
